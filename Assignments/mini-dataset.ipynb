```python
# Mini dataset generators (no external files)

def make_student_success(n=600, seed=1):
    rng = np.random.default_rng(seed)
    X = pd.DataFrame({
        "hours": rng.integers(0, 12, n),
        "attendance": rng.integers(40, 100, n),
        "sleep": rng.integers(4, 9, n),
        "practice_tests": rng.integers(0, 6, n),
    })
    y = (((X["hours"] >= 5) & (X["attendance"] >= 70) & (X["sleep"] >= 6)) | (X["practice_tests"] >= 4)).astype(int)
    y = np.where(rng.random(n) < 0.07, 1-y, y)  # noise
    return X, y

def make_housing(n=500, seed=2):
    rng = np.random.default_rng(seed)
    size = rng.normal(1600, 350, n)
    beds = rng.integers(1, 6, n)
    city = rng.choice(["Toronto","Montreal","Vancouver","Calgary"], size=n)
    base_price = 180*size + 35000*beds
    bump = pd.Series(city).map({"Toronto":200000,"Vancouver":250000,"Montreal":80000,"Calgary":60000}).to_numpy()
    price = base_price + bump + rng.normal(0, 30000, n)
    X = pd.DataFrame({"size_sqft": size, "beds": beds, "city": city})
    return X, price

def make_netflix(n=700, seed=3):
    rng = np.random.default_rng(seed)
    clicks = rng.integers(0, 15, n)
    weekend = rng.integers(0, 2, n)
    minutes = 18*clicks + 30*weekend + rng.normal(0, 15, n)
    binge = ((clicks >= 7) | (minutes >= 140)).astype(int)
    X = pd.DataFrame({"clicks": clicks, "is_weekend": weekend})
    return (X, minutes), (X.copy(), binge)

def make_healthcare(n=700, seed=4):
    rng = np.random.default_rng(seed)
    X = pd.DataFrame({
        "age": rng.integers(1, 90, n),
        "fever": rng.normal(37.0, 1.0, n),
        "oxygen": rng.normal(97, 2.5, n),
    })
    y = (((X["age"] > 60) & (X["fever"] > 38.0)) | (X["oxygen"] < 94)).astype(int)
    y = np.where(rng.random(n) < 0.06, 1-y, y)
    return X, y

def make_fraud(n=2000, seed=5):
    rng = np.random.default_rng(seed)
    X = pd.DataFrame({
        "amount": rng.normal(60, 25, n).clip(1, None),
        "is_international": rng.integers(0, 2, n),
        "night": rng.integers(0, 2, n),
    })
    y = ((X["amount"] > 95) & (X["is_international"] == 1) & (X["night"] == 1)).astype(int)
    y = np.where(rng.random(n) < 0.93, 0, y)  # rare
    return X, y
```
