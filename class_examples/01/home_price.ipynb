```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression  # Added import for LinearRegression

rng = np.random.default_rng(1)
size_sqft = rng.normal(1500, 300, 120)
price = 220 * size_sqft + rng.normal(0, 20000, 120)

#print(size_sqft)
#print(price)

df = pd.DataFrame({"size_sqft": size_sqft, "price": price})

#df
#df.head()

x = df[["size_sqft"]]
y = df[["price"]]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)

model = LinearRegression()  # Now this will work with the proper import
model.fit(x_train, y_train)

pred = model.predict(x_test)

new_house = [[1800]]
new_pred = model.predict(new_house)

print(new_pred[0]);
```
