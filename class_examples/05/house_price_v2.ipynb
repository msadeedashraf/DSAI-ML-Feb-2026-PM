```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

def make_housing(n=700, seed=205):
    rng = np.random.default_rng(seed)
    size = rng.normal(1600, 350, n)
    beds = rng.integers(1, 6, n)
    city = rng.choice(["Toronto","Montreal","Vancouver","Calgary"], size=n)

    base_price = 180*size + 35000*beds
    bump = pd.Series(city).map({
        "Toronto":200000,
        "Vancouver":250000,
        "Montreal":80000,
        "Calgary":60000
    }).to_numpy()

    price = base_price + bump + rng.normal(0, 30000, n)
    X = pd.DataFrame({"size_sqft": size, "beds": beds, "city": city})
    y = price
    return X, y

    # 1) Data
X, y = make_housing(n=700, seed=205)

X_num = X[["size_sqft", "beds"]]
# 2) Split
X_train, X_test, y_train, y_test = train_test_split(
    X_num, y, test_size=0.25, random_state=205
)

baseline = LinearRegression().fit(X_train[["size_sqft","beds"]], y_train)
pred_base = baseline.predict(X_test[["size_sqft","beds"]])


model = LinearRegression()


model.fit(X_train, y_train)
pred_full = model.predict(X_test)
mae_full = mean_absolute_error(y_test, pred_full)

print(mae_full)
```
