```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score



rng = np.random.default_rng(42)

n = 300

hours = rng.integers(0, 11, n)
attendance = rng.integers(50, 101, n)
practice_tests = rng.integers(0, 6, n)

X = pd.DataFrame({
    "hours": hours,
    "attendance": attendance,
    "practice_tests": practice_tests
})

y = (
    ((X["hours"] >= 5) & (X["attendance"] >= 75)) |
    (X["practice_tests"] >= 4)
).astype(int)


noise_idx = rng.choice(n, size=25, replace=False)
y.iloc[noise_idx] = 1 - y.iloc[noise_idx]
#X
#y

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)


depth = [1,3,8]

models = {}

for d in depth:
    model = DecisionTreeClassifier(max_depth=d, random_state= 42)
    model.fit(X_train, y_train)
    models[d] = model


print("Accuracy ")
print("-" * 40)

for d, model in models.items():
    train_acc = accuracy_score(y_train, model.predict(X_train))
    test_acc = accuracy_score(y_test, model.predict(X_test))
    print(f"max_depth{d} | Train Accuracy= {train_acc:.3f} | Test Accuracy = {test_acc:.3f}")


for d, model in models.items():
    plt.figure(figsize=(14,8))
    plot_tree(
        model,
        feature_names=X.columns,
        class_names=["Fail", "Pass"],
        filled = True,
        rounded=True,fontsize=10
    )
    plt.title(f"Decision Tree max_depth={d}")
    plt.show()
    
new_students = pd.DataFrame({
    "hours": [2, 6, 4],
    "attendance": [60, 85, 78],
    "practice_tests": [1, 2, 5]
})
    

best_model = models[3]
pred = best_model.predict(new_students)
print("\nPredictions using depth = 3")
print(new_students)
print(f"Predicted : {pred}")
```
