```python
import numpy as np
import pandas as pd


from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    mean_absolute_error, mean_squared_error, r2_score,
    precision_score, recall_score, f1_score
)


from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


def make_housing(n=700, seed=2):
    rng = np.random.default_rng(seed)
    size = rng.normal(1600, 350, n)
    beds = rng.integers(1, 6, n)
    city = rng.choice(["Toronto","Montreal","Vancouver","Calgary"], size=n)
    base_price = 180*size + 35000*beds
    bump = pd.Series(city).map({"Toronto":200000,"Vancouver":250000,"Montreal":80000,"Calgary":60000}).to_numpy()
    price = base_price + bump + rng.normal(0, 30000, n)
    X = pd.DataFrame({"size_sqft": size, "beds": beds, "city": city})
    return X, price


X,y = make_housing()

#X

X = pd.get_dummies(X) # Converting cities into numbers

#X

X_train, X_test, y_train, y_test = train_test_split(
            X, y,test_size=0.2,random_state=1
        )

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

#X_train
#X_test



model = KNeighborsRegressor(n_neighbors=5)
model.fit(X_train, y_train)


lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.score(X_test, y_test))

pred = model.predict(X_test)

#pred 


print(model.score(X_test, y_test))

#house = [[2000, 3,1,0,0,0]]
#house = [[2000, 3,0,0,0,1]]
#print("Prediction : " , model.predict(house)[0])

#print(model.score())


```
