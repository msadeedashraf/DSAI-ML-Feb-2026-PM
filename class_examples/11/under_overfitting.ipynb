```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


# Generate dataset with noise
np.random.seed(0)

hours = np.random.randint(1, 11, 40)

# Real rule (imperfect)
pass_exam = (hours >= 6).astype(int)

# Add noise (flip some labels)
noise_index = np.random.choice(len(pass_exam), 6, replace=False)
pass_exam[noise_index] = 1 - pass_exam[noise_index]

data = pd.DataFrame({
    "hours": hours,
    "pass": pass_exam
})

X = data[["hours"]]
y = data["pass"]


# Tarin/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


underfit_model = DecisionTreeClassifier(max_depth=1)
underfit_model.fit(X_train, y_train)


print("Underfit Train:", underfit_model.score(X_train, y_train))
print("Underfit Test:", underfit_model.score(X_test, y_test))



good_model = DecisionTreeClassifier(max_depth=3)
good_model.fit(X_train, y_train)

print("Good Train:", good_model.score(X_train, y_train))
print("Good Test:", good_model.score(X_test, y_test))
```
