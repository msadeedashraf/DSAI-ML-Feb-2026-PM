```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt


np.random.seed(0)

hours = np.linspace(1,10,50)
scores = 5*hours + np.random.normal(0,3,50)

data = pd.DataFrame({
    "hours":hours,
    "score":scores
})

X = data[["hours"]]
y = data["score"]


X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.25,random_state=42
)

underfit = DecisionTreeRegressor(max_depth=1)
underfit.fit(X_train,y_train)

print("Underfit Train:",underfit.score(X_train,y_train))
print("Underfit Test:",underfit.score(X_test,y_test))

good_model = DecisionTreeRegressor(max_depth=3)
good_model.fit(X_train,y_train)

print("Good Train:",good_model.score(X_train,y_train))
print("Good Test:",good_model.score(X_test,y_test))


overfit = DecisionTreeRegressor()
overfit.fit(X_train,y_train)

print("Overfit Train:",overfit.score(X_train,y_train))
print("Overfit Test:",overfit.score(X_test,y_test))

x_range = np.linspace(X.hours.min(), X.hours.max(), 200).reshape(-1,1)

plt.scatter(X,y,label="Data")

plt.plot(x_range, underfit.predict(x_range),label="Underfit")
plt.plot(x_range, good_model.predict(x_range),label="Good Fit")
plt.plot(x_range, overfit.predict(x_range),label="Overfit")

plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```
