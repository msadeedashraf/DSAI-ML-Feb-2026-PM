```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer


data = pd.DataFrame({
    "text": [
        "Win a free iPhone now",
        "Lowest price deal click here",
        "Hey are we meeting tomorrow?",
        "Can you send the notes?",
        "Free cash prize claim now",
        "Lunch at 1 pm?"
    ],
    "label": [1,1,0,0,1,0]  # 1=spam, 0=not spam
})


X_train, X_test, y_train, y_test = train_test_split(data["text"], data["label"], test_size=0.33, random_state=3)

vectorizer = CountVectorizer()
X_train_vec =  vectorizer.fit_transform(X_train)
X_test_vec  =  vectorizer.transform(X_test)

model = LogisticRegression()
model.fit(X_train_vec, y_train)

pred = model.predict(X_test_vec)

msg = ["Lunch at 1 pm?"]

msg_vec  =  vectorizer.transform(msg)

print(model.predict(msg_vec)[0])


```
