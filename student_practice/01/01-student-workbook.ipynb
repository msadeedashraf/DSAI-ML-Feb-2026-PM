## Reusable Mini-Datasets
These functions generate small, relatable datasets for examples.

```python
# ✅ generate small datasets quickly (no external files)

def make_student_success(n=500, seed=1):
    rng = np.random.default_rng(seed)
    hours = rng.integers(0, 12, n)
    attendance = rng.integers(40, 100, n)
    sleep = rng.integers(4, 9, n)
    practice_tests = rng.integers(0, 6, n)
    y = (((hours >= 5) & (attendance >= 70) & (sleep >= 6)) | (practice_tests >= 4)).astype(int)
    y = np.where(rng.random(n) < 0.07, 1-y, y)  # a bit of noise
    X = pd.DataFrame({"hours": hours, "attendance": attendance, "sleep": sleep, "practice_tests": practice_tests})
    return X, y

def make_housing(n=400, seed=2):
    rng = np.random.default_rng(seed)
    size = rng.normal(1600, 350, n)
    beds = rng.integers(1, 6, n)
    city = rng.choice(["Toronto","Montreal","Vancouver","Calgary"], size=n, replace=True)
    base = 180*size + 35000*beds
    city_bump = pd.Series(city).map({"Toronto":200000,"Vancouver":250000,"Montreal":80000,"Calgary":60000}).to_numpy()
    price = base + city_bump + rng.normal(0, 30000, n)
    X = pd.DataFrame({"size_sqft": size, "beds": beds, "city": city})
    y = price
    return X, y

def make_netflix(n=600, seed=3):
    rng = np.random.default_rng(seed)
    clicks = rng.integers(0, 15, n)
    is_weekend = rng.integers(0, 2, n)
    watch_minutes = 18*clicks + 30*is_weekend + rng.normal(0, 15, n)
    binge = ((clicks >= 7) | (watch_minutes >= 140)).astype(int)
    Xr = pd.DataFrame({"clicks": clicks, "is_weekend": is_weekend})
    yr = watch_minutes
    Xc = Xr.copy()
    yc = binge
    return (Xr, yr), (Xc, yc)

def make_healthcare(n=600, seed=4):
    rng = np.random.default_rng(seed)
    age = rng.integers(1, 90, n)
    fever = rng.normal(37.0, 1.0, n)
    oxygen = rng.normal(97, 2.5, n)
    high_risk = (((age > 60) & (fever > 38.0)) | (oxygen < 94)).astype(int)
    high_risk = np.where(rng.random(n) < 0.06, 1-high_risk, high_risk)
    X = pd.DataFrame({"age": age, "fever": fever, "oxygen": oxygen})
    y = high_risk
    return X, y

def make_fraud(n=1200, seed=5):
    rng = np.random.default_rng(seed)
    amount = rng.normal(60, 25, n).clip(1, None)
    is_international = rng.integers(0, 2, n)
    night = rng.integers(0, 2, n)
    fraud = ((amount > 95) & (is_international == 1) & (night == 1)).astype(int)
    # make fraud rare
    fraud = np.where(rng.random(n) < 0.93, 0, fraud)
    X = pd.DataFrame({"amount": amount, "is_international": is_international, "night": night})
    y = fraud
    return X, y
```
